# Diagnosis eda:


In this dataset th eattribute 'diagnosis' is the target variable. 

```python
import seaborn as sns
```

```python
sns.catplot(x='diagnosis',data=data,kind='count')
```

Most of the observations in this dataset is found with a benign diagnosis,whereas less values of malignant have been observed.

```python
dataA=data.drop(["id"],axis=1)
```

```python
for i in dataA:
    if i != 'diagnosis':
        sns.catplot(x='diagnosis',y=i,data=dataA,kind='box')
```

From all of these box plot it is clear that mostof the values are usually higher in malignant than that of benign

```python
for i in dataA:
    if i != 'diagnosis':
        sns.catplot(x='diagnosis',y=i,data=dataA,kind='swarm')
```

The category malignant has a good spred of data,whereas th ecategory benign has a muchhigher count.
